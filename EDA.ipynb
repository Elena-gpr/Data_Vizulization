```python
# Libraries 
import pandas as pd
import numpy as np
```

### Load Data 

```python
data = pd.read_csv('cost-of-living_v2.csv')	
data.head()
```

```python
data.info() # Total data 4956
```

```python
data.isnull().sum().reset_index()
```

### Data Cleaning

```python
# Delete columns
col = ['x1', 'x2', 'x4', 'x5', 'x6',
       'x7', 'x8', 'x24', 'x25', 'x26', 
       'x27', 'x28', 'x29', 'x30', 'x31', 
       'x32', 'x34', 'x35', 'x37', 'x40',
       'x42' , 'x43', 'x44', 'x45', 'x46',
       'data_quality']

df = data.drop(col, axis='columns')
df.head()
```

```python
# Total null valor by row
df['nulos_por_fila'] = df.isnull().sum(axis=1)
df.shape
```

```python

# All rows with more than 10 null columns are eliminated
df = df[df['nulos_por_fila'] <= 10]
```

```python
df.isnull().sum()
```

```python
df[df['x3'].isnull() == True ]
```

```python
mean_by_country = df.groupby('country').transform('mean')


# Replace null values ​​with the mean for its country
df_filled = df.copy()
for column in mean_by_country.columns:
    df_filled[column].fillna(mean_by_country[column], inplace=True)

df_filled
```

```python
df_filled.isnull().sum()
```

```python
#Delete null values 
df_filled.drop('nulos_por_fila', axis=1, inplace=True)
```

```python
df_filled.dropna(inplace=True)

df_filled.isnull().sum()
```

```python
# salary

data_salary = pd.read_csv('salary_data.csv')
data_salary.head()
```

```python
data_salary['country_name'].unique()
```

```python
data_salary.info()
```

```python
# Merge with data filled
df_join_salary = df_filled.merge(data_salary, how='left', left_on='country',
                                 right_on='country_name')

df_join_salary.head()
```

```python

df_join_salary[df_join_salary['city'].duplicated() == True]
```

```python
index = pd.read_csv('Cost_of_Living_Index_2022.csv')
index.head()
```

```python
df_join_index = df_join_salary.merge(index, how='left', left_on='country',
                                     right_on='Country')

df_join_index.head()
```
